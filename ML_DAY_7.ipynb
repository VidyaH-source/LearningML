```python
#label, and title
xpoints = [1,2,3,4,5]
ypoints = [10,20,30,40,50]

plt.plot(xpoints, ypoints)
plt.xlabel("day")
plt.ylabel("sale")
plt.title("Sale per day")
plt.show()
```

```python
# Setting font properties for label, and title

xpoints = [1,2,3,4,5]
ypoints = [10,20,30,40,50]

plt.plot(xpoints, ypoints)

font1 ={'family':'serif', 'color':'blue', 'size':18}
font2 ={'family':'serif', 'color':'darkred', 'size':15}

plt.title("Sale per day",fontdict = font1)
plt.xlabel("Day",fontdict = font2)
plt.ylabel("Sale",fontdict = font2)

plt.show()
```

```python
# Setting font properties for label, and title
#loc means location, 3 values are right , left, center
xpoints = [1,2,3,4,5]
ypoints = [10,20,30,40,50]

plt.plot(xpoints, ypoints)

font1 ={'family':'serif', 'color':'blue', 'size':18}
font2 ={'family':'serif', 'color':'darkred', 'size':15}

plt.title("Sale per day",fontdict = font1, loc = 'right')
plt.xlabel("Day",fontdict = font2)
plt.ylabel("Sale",fontdict = font2)

plt.show()
```

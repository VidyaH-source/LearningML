```python
import matplotlib.pyplot as plt

#Multiple points
xpoints = [1,2,3,4,5]
ypoints = [10,20,30,40,50]
plt.plot(xpoints,ypoints)
plt.show()
```

```python
#2 points
xpoints = [0,6]
ypoints = [0,250]
plt.plot(xpoints,ypoints)
plt.show()
```

```python
#default x points
#always take y.. default for x values will start from [0,1,2...]

xpoints = [10,20,30,40,50,60]
plt.plot(xpoints)
plt.show()
```

```python
# format strings fmt    marker|line|color

ypoints = [10,20,30,40,50,60]
plt.plot(ypoints, marker = '*')
plt.show()
```

```python
ypoints = [10,20,30,40,50,60]
plt.plot(ypoints, marker = 'o')
plt.show()
```

```python
ypoints = [10,20,30,40,50,60]
plt.plot(ypoints, marker = 'v')
plt.show()
```

```python
ypoints = [10,20,30,40,50,60]
plt.plot(ypoints, marker = 'v', ms = 15) # ms is marker size
plt.show()
```

```python
ypoints = [10,20,30,40,50,60]
plt.plot(ypoints, marker = 'v', ms = 10, color = 'g') # color
plt.show()
```

```python
ypoints = [10,20,30,40,50,60]
# mfc is marker face colour
plt.plot(ypoints, marker = 'v', ms = 10, color = 'g', mfc = 'b')
plt.show()
```

```python
ypoints = [10,20,30,40,50,60]
plt.plot(ypoints, 'o:k') #o is marker, : is dotted line, k is black color
plt.show()
```

```python
""" '-' solid line
':' dotted line
'--' dashed line
'-.' dash-dot line"""
ypoints = [10,20,30,40,50,60]
plt.plot(ypoints, 'o--k')
plt.show()
```

```python
""" '-' solid line
':' dotted line
'--' dashed line
'-.' dash-dot line"""
ypoints = [10,20,30,40,50,60]
plt.plot(ypoints, 'o-.k',linewidth ='10')
plt.show()
```
